```python
"""
Задание: 52 (qwrqwr: 2010-04-23)
Определить названия всех кораблей из таблицы Ships, которые могут быть линейным японским кораблем,
имеющим число главных орудий не менее девяти, калибр орудий менее 19 дюймов 
и водоизмещение не более 65 тыс.тонн
"""

sql = """
select distinct name from (
  select case when 
  (type = 'bb' or type is null) and 
  (country = 'japan' or country is null) and 
  (numGuns >= 9 or numGuns is null) and
  (bore < 19.0 or bore is null) and
  (displacement <= 65000 or displacement is null)
  then name end as name
  from (
    select name, type, country, numGuns, bore, displacement
    from classes c 
    join ships s on c.class = s.class
  ) t
) t2 where name is not null
"""
```

```python
"""
Задание: 53 (Serge I: 2002-11-05)
Определите среднее число орудий для классов линейных кораблей.
Получить результат с точностью до 2-х десятичных знаков.
"""

sql = """
select 
cast((sum(numGuns) / (count(*) * 1.0)) AS NUMERIC(6,2)) as avg_numGuns
from classes
where type = 'bb'
group by type
"""
```

```python
"""
Задание: 54 (Serge I: 2003-02-14)
С точностью до 2-х десятичных знаков определите 
среднее число орудий всех линейных кораблей (учесть корабли из таблицы Outcomes).
"""

sql = """
select
cast(avg(avg_numGuns) as decimal(6,2)) as avg_numGuns
from (
  select c.type, s.name, cast(c.numGuns as decimal(6,2)) as avg_numGuns
  from classes c 
  join ships s on c.class = s.class
  union
  select c.type, o.ship, cast(c.numGuns as decimal(6,2)) as avg_numGuns
  from classes c 
  join outcomes o on c.class = o.ship
) t
where type = 'bb'
group by type
"""
```

```python
"""
Задание: 55 (Serge I: 2003-02-16)
Для каждого класса определите год, когда был спущен на воду первый корабль этого класса. 
Если год спуска на воду головного корабля неизвестен, определите минимальный 
год спуска на воду кораблей этого класса. 
Вывести: класс, год.
"""

sql = """
select c.class, min(s.launched) as year
from classes c
left join ships s on c.class = s.class
group by c.class

"""
```

```python
"""
Задание: 56 (Serge I: 2003-02-16)
Для каждого класса определите число кораблей этого класса, потопленных в сражениях. 
Вывести: класс и число потопленных кораблей.
"""

sql = """
select c.class, count(name) as Sunks
from classes c
left join (
  select c.class, name
  from classes c
  join ships s on c.class = s.class
  where name in (select ship from outcomes where result = 'sunk')
  union
  select class, ship 
  from classes c 
  join outcomes o on c.class = o.ship
  where result = 'sunk'
) t on c.class = t.class
group by c.class

"""
```

```python
"""
Задание: 57 (Serge I: 2003-02-14)
Для классов, имеющих потери в виде потопленных кораблей и не менее 3 кораблей в базе данных, 
вывести имя класса и число потопленных кораблей.
"""

sql = """
with t as (
  select tt.class, count(tt.name) as cnt
  from (
    select c.class, s.name as name
    from classes c
    join ships s on c.class = s.class
    where s.name in (
      select o.ship as name from outcomes o where o.result = 'sunk')
    union
    select c.class, o.ship as name
    from classes c
    join outcomes o on c.class = o.ship
    where o.result = 'sunk'
  ) tt
  group by tt.class
),

t1 as (
  select tt1.class, count(tt1.name) as cnt
  from (
    select c.class, s.name as name
    from classes c
    join ships s on c.class = s.class
    union
    select c.class, o.ship as name
    from classes c
    join outcomes o on c.class = o.ship
  ) tt1
  group by tt1.class
  having count(tt1.name) >= 3
)

select t.class, t.cnt
from t
join t1 on t.class = t1.class
"""
```
